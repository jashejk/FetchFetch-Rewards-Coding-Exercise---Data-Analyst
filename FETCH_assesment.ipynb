# Brands data

```python
import json
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
data=[]
with open('brands.json', 'r') as file:
    for line in file:
        data.append(json.loads(line))
```

# Brands Logic

```python
id= [i['_id']['$oid'] if '_id' in i.keys() else ''  for i in data]
barcode= [i['barcode'] if 'barcode' in i.keys() else ''  for i in data]
category= [i['category'] if 'category' in i.keys() else ''  for i in data]
categoryCode= [i['categoryCode'] if 'categoryCode' in i.keys() else '' for i in data]
name= [i['name'] if 'name' in i.keys() else '' for i in data]
topBrand= [i['topBrand'] if 'topBrand' in i.keys() else '' for i in data]
brandCode= [i['brandCode'] if 'brandCode' in i.keys() else ''  for i in data] 
cpg_id= [i['cpg']['$id']['$oid'] if '$id' in i['cpg'].keys() else ''  for i in data]   
ref= [i['cpg']['$ref'] if '$ref' in i['cpg'].keys() else ''  for i in data]   
```

```python
df_brands=pd.DataFrame({'id':id,'barcode':barcode,'category':category,
             'categoryCode':categoryCode,'name':name,'topBrand':topBrand,
             'brandCode':brandCode,'cpg_id':cpg_id,'ref':ref})
```

```python
df_brands
```

```python
same_values = df_brands[df_brands['barcode'] == df_brands['brandCode']]

print("Records where barcode and brandCode have the same values:")
print(len(same_values))
```

```python
# Generate the profiling report
profile = ProfileReport(df_brands)
```

```python
profile.to_file("brandsdata_report.html")
```

# Receipts

```python
rec_data=[]
with open('receipts.json', 'r') as file:
    for line in file:
        rec_data.append(json.loads(line))
```

```python
df_rec=pd.DataFrame(rec_data)
```

```python
df_rec['_id']=df_rec['_id'].apply(pd.Series)
```

```python
df_rec['createDate']=df_rec['createDate'].apply(pd.Series)["$date"]
df_rec['dateScanned']=df_rec['dateScanned'].apply(pd.Series)["$date"]
df_rec['finishedDate']=df_rec['finishedDate'].apply(pd.Series)["$date"]
df_rec['modifyDate']=df_rec['modifyDate'].apply(pd.Series)["$date"]
df_rec['pointsAwardedDate']=df_rec['pointsAwardedDate'].apply(pd.Series)["$date"]
df_rec['purchaseDate']=df_rec['purchaseDate'].apply(pd.Series)["$date"]
```

```python
df_rec.info()
```

```python
# Generate the profiling report
profile = ProfileReport(df_rec)
profile.to_file("receiptsdata_report.html")
```

```python
items=[]
for i in zip(df_rec['_id'],df_rec['rewardsReceiptItemList'].values):
    if type(i[1])==type([]):
        for j in i[1]:
            j['_id']=i[0]
            items.append(j)
    
```

```python
df_items=pd.DataFrame(items)
```

```python
df_items.columns.values
```

```python
df_items
```

```python
df_items["rewardsProductPartnerId"].value_counts()
```

```python
df_receipts=df_rec.merge(df_items,on="_id",how='inner')
```

```python
df_br_rc=df_brands.merge(df_receipts,left_on="cpg_id",right_on="rewardsProductPartnerId",how='inner')
```
